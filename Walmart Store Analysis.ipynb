## If the weekly sales are affected by the unemployment rate, if yes - which stores are suffering the most?

```python
stores = list(data['Store'].unique())
correlation_unemployment = dict()
for i in stores:
    df = data[data['Store'] == i]
    correlation_unemployment[i] = df['Unemployment'].corr(df['Weekly_Sales'])
    
df_corr = pd.DataFrame.from_dict(correlation_unemployment, orient='index', columns=['Correlation'])

# Turn the index into a column
df_corr = df_corr.reset_index().rename(columns={'index': 'Store'})
df_corr.sort_values(by = 'Correlation')
```

* Above shows correlation result between Unemployment rate and Weekly Sales for each Store.
* For some Stores sales increase as unemployment increases (possibly unusual or due to other local effects)
* For It can be seen that as the unemployment rate increases sales decreases (this is a more usual trend). A negative correlation is observed.
* Store 38, 44, 39, 42, 41 are the most affected stores who are suffering due to increasing unemployment rate.

# Use predictive modeling techniques to forecast the sales for each store for the next 12weeks.

```python
from prophet import Prophet
```

```python
data.info()
```

```python
store_forecasts = {} ##Dictonary to store forecasting data of each store

stores = list(data['Store'].unique())
for store in stores:
    store_df = data[data['Store'] == store][['Date', 'Weekly_Sales']]
    store_df.rename(columns = {'Date' : 'ds', 'Weekly_Sales' : 'y'}, inplace = True)
        
    model = Prophet()
    model.fit(store_df)
    
    future_pred = model.make_future_dataframe(periods = 12, freq = 'W')
    
    forecast = model.predict(future_pred)
    forecast_12 = forecast[['ds','yhat']].tail(12)
    
    store_forecasts[store] = forecast_12
    print(store_forecasts[store])
    
```

```python
merged_stores = {}
for store_id in stores:
    actual_df = globals()[f'Store{store_id}'][['Date', 'Weekly_Sales']].copy()
    
    actual_df.rename(columns = {'Date' : 'ds', 'Weekly_Sales' : 'Actual Sales'}, inplace = True)
    actual_df['ds'] = pd.to_datetime(actual_df['ds'])
    
    if store_id in store_forecasts:
        forecast_df = store_forecasts[store_id][['ds', 'yhat']].rename(columns = {'yhat' : 'Forecasted Sales'})
        
        merged_df = pd.merge(actual_df, forecast_df, on='ds', how = 'outer')
        
        merged_stores[store_id] = merged_df
```

```python

for i in stores:
    plt.figure(figsize = (8,5))
    sns.lineplot(x = 'ds', y='Actual Sales', data = merged_stores[i], color = 'Blue')
    sns.lineplot(x = 'ds', y='Forecasted Sales', data = merged_stores[i], color = 'Red')
    plt.show()
```
